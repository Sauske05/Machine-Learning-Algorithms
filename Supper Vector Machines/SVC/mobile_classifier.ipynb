```python
import pandas as pd #type: ignore
df = pd.read_csv('mobile_data/train.csv')
df
```

```python
input_features = df.columns[:-1]
output_features = df.columns[-1]
X_train = df[input_features]
y_train = df[output_features]
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
parameters = {
    'kernel' : ('rbf', 'linear'), 'C': [0.1 ,1, 10, 100]
}
svc_model = SVC()
tuner_model = GridSearchCV(svc_model, parameters)
tuner_model.fit(X_train, y_train)
```

```python
test_df = pd.read_csv('mobile_data/test.csv')
test_df
```
